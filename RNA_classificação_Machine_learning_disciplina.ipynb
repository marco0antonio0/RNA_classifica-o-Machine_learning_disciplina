# Aplicação de RNA para fazer a predição de chances de ir no show

**Problematica de classificação**

Nesta aplicação temos a predição de um certo personagem ficticio que esta pensando em ir no show porem esta usando esta aplicação para avaliar as chances de ir ou não no show por meio de uma rede neural artificial utilizando da biblioteca tensorflow para criar a RNA para **classificar** as chances de ir ao show\
Foi utilizado como parametros para o modelo: \
**Modelo de treinamento**
1. camada para os dados de entrada
  - Função de ativação relu
  - Entrada de 3 valores
  - 10 Neuronios

2. camada de saida
  - Função de ativação sigmoid
  - Saida de valores binarios '0' ou '1'
  - 1 Neuronio

3. otimizadores do keras
  - adam

Entradas:

| Tem Amigos | Show é Longe | Ingresso é Caro |
|------------|--------------|-----------------|

\


## Imports necessarios
> tensorflow \
> numpy \
> matplotlib \

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from matplotlib import pyplot
```

## Preparação dos Dados
#### Descrição da tabela
**Array x**
\
coluna 1 - tem amigos \
coluna 2 - show é longe \
coluna 3 - ingresso é caro \

| Tem Amigos | Show é Longe | Ingresso é Caro |
|------------|--------------|-----------------|


**Array y**
\
Resultado do array x 1 para verdadeiro e 0 para falso

| Decisão (Ir ao Show?) |
|-----------------------|





### Explicação
Nesta etapa é preparado os dados em que o array x é atribuido o conjunto de arrays a matriz referente aos dados de treinamento \

Sendo o array x os dados de treinamento e o array y os dados de indexação referente ao array x cujo se refere aos dados resultantes verdadeiro da logica em que para o valor ser verdadeiro é necessario que para ir ao show precisa ter amigos

```python
#======================================
# Dados de entrada
# descrição das colunas
# [tem amigos, show é longe, ingresso é caro]
x = np.array([
    [0, 0, 0],  # resultado ==> 0 | descrição: Não tem amigos, show perto e ingresso barato
    [0, 0, 1],  # resultado ==> 0 | descrição: Não tem amigos, show perto e ingresso caro
    [0, 1, 0],  # resultado ==> 0 | descrição: Não tem amigos, show longe e ingresso barato
    [0, 1, 1],  # resultado ==> 0 | descrição: Não tem amigos, show longe e ingresso caro
    [1, 0, 0],  # resultado ==> 1 | descrição: Tem amigos, show perto e ingresso barato
    [1, 0, 1],  # resultado ==> 1 | descrição: Tem amigos, show perto e ingresso caro
    [1, 1, 0],  # resultado ==> 1 | descrição: Tem amigos, show longe e ingresso barato
    [1, 1, 1]]) # resultado ==> 1 | descrição: Tem amigos, show longe e ingresso caro


# Saída (irá ao show? 1 para sim, 0 para não)
# A regra é: precisa ter amigos para ir ao show
y = np.array([0,0,0,0,1,1,1,1])  # A decisão depende apenas de ter amigos, conforme a regra
```

## Construção do Modelo
---

**Descrições**


---

**mod.add(Dense(10, activation='relu', input_shape=(3,))):** \
Esta linha adiciona uma camada densa à rede neural. Uma camada densa é uma camada onde cada neurônio na camada anterior está conectado a cada neurônio na camada seguinte. Neste caso, estamos adicionando uma camada densa com 10 neurônios (unidades).\
A função de ativação usada é a **'relu'** (Rectified Linear Unit), que é uma função de ativação comumente usada em redes neurais. \
\
**input_shape=(3,)** \
Especifica a forma dos dados de entrada. Aqui, temos 3 features, então o formato de entrada é (3,). \
Esta é a primeira camada da rede, portanto, precisamos especificar o formato da entrada.


| Tem Amigos | Show é Longe | Ingresso é Caro |
|------------|--------------|-----------------|


---


\
**mod.add(Dense(1, activation='sigmoid')):** \
Esta linha adiciona outra camada densa à rede neural, desta vez com apenas 1 neurônio. A função de ativação usada é **'sigmoid'**. \
A função sigmoid é comumente usada em problemas de classificação binária, pois mapeia a saída para um intervalo entre 0 e 1, representando a probabilidade de a saída ser da classe positiva. \
Neste caso, estamos lidando com uma classificação binária (ir ou não ir ao show), então usamos a função de ativação sigmoid na camada de saída.

### Explicação
Nesta etapa criamos o modelo de treinamento
1. camada para os dados de entrada
  - Função de ativação relu
  - Entrada de 3 valores
  - 10 Neuronios

2. camada de saida
  - Função de ativação sigmoid
  - Saida de valores binarios '0' ou '1'
  - 1 Neuronio

3. otimizadores do keras
  - adam

```python
mod = Sequential()
mod.add(Dense(10, activation='relu', input_shape=(3,)))
mod.add(Dense(1, activation='sigmoid'))
mod.compile(optimizer='adam',
            loss='binary_crossentropy',
            metrics=['accuracy'])
```

## Treinamento do Modelo
x = dados utilizados para treinamento \
y = index de resultados do array x

### Explicação
Nesta etapa estamos realizando o treinamento com o conjunto de dados que foram preparados nas etapas anteriores utilizando o modelo de treinamento que foi preparado anteriomente e temos o treinamento realizado em determinada quantidades de epocas no caso 500

```python
# Treinamento do modelo e obtenção do histórico
history = mod.fit(x, y, epochs=500)
```

## Gerar grafico

Extrair precisão de treinamento do histórico

Plotar gráfico

### Explicação
Nesta etapa demostramos por meio do grafico em qual momento o modelo de treinamento convergiu em relação as epocas totais que foram de 500

```python
# Extrair precisão de treinamento do histórico
accuracy = history.history['accuracy']
# Plotar gráfico
pyplot.plot(range(1, len(accuracy) + 1), accuracy)
pyplot.xlabel('Épocas')
pyplot.ylabel('Precisão')
pyplot.title('Precisão do Modelo ao Longo das Épocas')
pyplot.grid(True)
pyplot.show()
```

## Avaliação
Avaliando o modelo com os mesmos dados, em um caso real, você deveria usar um conjunto de teste separado.

```python
mod.evaluate(x, y)
```

## Fazendo Predições
Exemplo: \
[1, 0, 0]  ===> Tem amigos, show perto e ingresso barato

### Explicação
Nesta etapa realizamos o teste da nossa aplicação de RNA com os seguintes dados: \
[1, 0, 0] \
Tem amigos, show perto e ingresso barato

```python
# Testando novos dados
# Exemplo: Uma pessoa com amigos, para um show perto e com ingresso barato
new_data = np.array([
    [1, 0, 0]  # Tem amigos, show perto e ingresso barato
])

prediction = mod.predict(new_data)
print(f'Probabilidade de ir ao show: {prediction[0][0]}')  # Quanto mais próximo de 1, maior a probabilidade de ir
```

## Fazendo Predições
Exemplo: \
[0, 0, 0]  ===> Não amigos,  show perto e ingresso barato

### Explicação
Nesta etapa realizamos o teste da nossa aplicação de RNA com os seguintes dados: \
[0, 1, 1] \
Não tem amigos, show é longe e ingresso é caro

```python
# Testando novos dados
# Exemplo: Uma pessoa com amigos, para um show perto e com ingresso barato
new_data = np.array([
    [0, 1, 1]  # Não tem amigos, show é longe e ingresso é caro
])

prediction = mod.predict(new_data)
print(f'Probabilidade de ir ao show: {prediction[0][0]}')  # Quanto mais próximo de 1, maior a probabilidade de ir
```
